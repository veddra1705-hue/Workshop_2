Cabin Feature

```python
train = train.drop(['Cabin'], axis=1, errors='ignore')
test = test.drop(['Cabin'], axis=1, errors='ignore')
```

Embarked Feature

```python
print(train['Embarked'].value_counts())

train['Embarked'].fillna(train['Embarked'].mode()[0], inplace=True)
test['Embarked'].fillna(test['Embarked'].mode()[0], inplace=True)
```

Age Feature (Fill missing values)

```python
train['Age'].fillna(train['Age'].median(), inplace=True)
test['Age'].fillna(test['Age'].median(), inplace=True)
```

```python
combine = [train, test]
```

```python
for dataset in combine:
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
pd.crosstab(train['Title'], train['Sex'])
```

The loop goes through each dataset and extracts titles (like Mr, Miss, etc.) from the Name column into a new column called “Title”.

```python
for dataset in combine:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Capt', 'Col',
    'Don', 'Dr', 'Major', 'Rev', 'Jonkheer', 'Dona'], 'Rare')
    dataset['Title'] = dataset['Title'].replace(['Countess', 'Lady', 'Sir'], 'Royal')
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')

train[['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

```python
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Royal": 5, "Rare": 6}
for dataset in combine:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)
train.head()
```

```python
train['Age'].fillna(train['Age'].median(), inplace=True)
test['Age'].fillna(test['Age'].median(), inplace=True)
```

This code fills missing values in the Age column with the median age.

```python
train = train.drop(['Name'], axis=1)
test = test.drop(['Name'], axis=1)
```

```python
train['Sex'] = train['Sex'].map({'male': 0, 'female': 1})
test['Sex'] = test['Sex'].map({'male': 0, 'female': 1})
```

This code converts gender (Sex) from text into numbers.
Male is replaced with 0 and female with 1.

```python
train['Embarked'] = train['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
test['Embarked'] = test['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
test['Fare'].fillna(test['Fare'].median(), inplace=True)

train['FareBand'] = pd.qcut(train['Fare'], 4, labels=[1,2,3,4])
test['FareBand'] = pd.qcut(test['Fare'], 4, labels=[1,2,3,4])

train = train.drop(['Fare'], axis=1)
test = test.drop(['Fare'], axis=1)
```

Missing values in Fare are filled using the median fare.
Fare is divided into 4 bands and labeled 1–4 to simplify the data.
The Fare column is dropped since we now use FareBand.

```python
train.head()
test.head()
train.isnull().sum()
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
train = train.drop(['Ticket'], axis=1)
```

```python
X = train.drop(['Survived', 'PassengerId'], axis=1)
y = train['Survived']

x_train, x_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# 1. Support Vector Machine
svm = SVC()
svm.fit(x_train, y_train)
svm_pred = svm.predict(x_val)
print("SVM Accuracy:", accuracy_score(y_val, svm_pred) * 100)

# 2. Decision Tree
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
dt_pred = dt.predict(x_val)
print("Decision Tree Accuracy:", accuracy_score(y_val, dt_pred) * 100)

# 3. Random Forest
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_val)
print("Random Forest Accuracy:", accuracy_score(y_val, rf_pred) * 100)

# 4. KNN
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
knn_pred = knn.predict(x_val)
print("KNN Accuracy:", accuracy_score(y_val, knn_pred) * 100)
```

In this step, different machine learning models such as Support Vector Machine (SVM), Decision Tree, Random Forest, and K-Nearest Neighbors (KNN) are created.

Each model is trained using the training data (x_train, y_train) and then tested on validation data (x_val) to make predictions.

Finally, the accuracy of each model is calculated and printed to compare their performance and identify the best model for the dataset.
